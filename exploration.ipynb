```python
import os
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
```

```python
# the chest xray dataset is split into train, test and val folders
data_dir = 'data/chest_xray'

for split in ['train', 'test', 'val']:
    normal = len(os.listdir(f'{data_dir}/{split}/NORMAL'))
    pneumonia = len(os.listdir(f'{data_dir}/{split}/PNEUMONIA'))
    print(f'{split}: {normal} normal, {pneumonia} pneumonia, total: {normal + pneumonia}')
```

```python
fig, axes = plt.subplots(2, 4, figsize=(12, 6))

# show 4 normal X-rays
normal_dir = f'{data_dir}/train/NORMAL'
for i, img_name in enumerate(os.listdir(normal_dir)[:4]):
    img = Image.open(f'{normal_dir}/{img_name}')
    axes[0, i].imshow(img, cmap='gray')
    axes[0, i].set_title('Normal')
    axes[0, i].axis('off')

# show 4 pneumonia X-rays
pneumonia_dir = f'{data_dir}/train/PNEUMONIA'
for i, img_name in enumerate(os.listdir(pneumonia_dir)[:4]):
    img = Image.open(f'{pneumonia_dir}/{img_name}')
    axes[1, i].imshow(img, cmap='gray')
    axes[1, i].set_title('Pneumonia')
    axes[1, i].axis('off')

plt.suptitle('Normal vs Pneumonia X-rays')
plt.tight_layout()
plt.show()
```

```python
# visualize class distribution
splits = ['Train', 'Test', 'Val']
normal_counts = [1341, 234, 8]
pneumonia_counts = [3875, 390, 8]

x = np.arange(len(splits))
width = 0.35

fig, ax = plt.subplots(figsize=(8, 5))
ax.bar(x - width/2, normal_counts, width, label='Normal', color='steelblue')
ax.bar(x + width/2, pneumonia_counts, width, label='Pneumonia', color='tomato')

ax.set_title('Dataset Distribution')
ax.set_xticks(x)
ax.set_xticklabels(splits)
ax.set_ylabel('Number of Images')
ax.legend()
plt.show()
```
